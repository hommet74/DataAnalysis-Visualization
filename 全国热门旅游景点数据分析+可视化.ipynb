# 3 Pyecharts数据可视化

## 3.1 销量前20热门景点数据

```python
#线性渐变 
color_js="""new echarts.graphic.LinearGradient(0,0,1,0,
      [{offset:0,color:'#009ad6'},{offset:1, color:'#ed1952'}],false)"""

p1=(Bar()
   .add_xaxis(sort_info.名称[:20].values.tolist())
   .add_yaxis('热门景点销量',sort_info.销量[:20].values.tolist(),itemstyle_opts=opts.ItemStyleOpts(color=JsCode(color_js)))
   .reversal_axis()
   .set_global_opts(title_opts=opts.TitleOpts(title='热门景点销量数据'),
                   yaxis_opts=opts.AxisOpts(name='景点名称'),
                   xaxis_opts=opts.AxisOpts(name='销量'),)
   .set_series_opts(label_opts=opts.LabelOpts(position='right')))

g1=(Grid()
   .add(p1,grid_opts=opts.GridOpts(pos_left='20%')))

g1.render_notebook()
```

## 3.2 假期出行数据全国地图分布

```python
p2=(Map()
   .add('假期出行分布',[list(z) for z in zip(df.groupby('城市').agg(销量=('销量','sum')).index.tolist(),df.groupby('城市').agg(销量=('销量','sum')).销量.values.tolist())],'china')
   .set_global_opts(title_opts=opts.TitleOpts(title='假期出行数据地图分布'),
                   visualmap_opts=opts.VisualMapOpts(max_=100000,range_color=['white','#fa8072','#ed1941'])))

p2.render_notebook()
```

## 3.3各省市4A-5A景区数量柱状图


```python
df_tmp=df[df['星级'].isin(['4A','5A'])]
df_tmp
```

```python
df_counts=df_tmp.groupby('城市').count()['星级']
df_counts
```

```python
p3=(Bar()
   .add_xaxis(df_counts.index.tolist())
   .add_yaxis('4A-5A景区数量',df_counts.values.tolist(),itemstyle_opts=opts.ItemStyleOpts(color=JsCode(color_js)),label_opts=opts.LabelOpts(is_show=True))
   .set_global_opts(title_opts=opts.TitleOpts(title='各省市4A-5A景区数量'),
                   datazoom_opts=opts.DataZoomOpts(is_show=True),
                   tooltip_opts=opts.TooltipOpts(is_show=True,trigger='axis',axis_pointer_type='cross'),
                   xaxis_opts=opts.AxisOpts(type_='category',axispointer_opts=opts.AxisPointerOpts(is_show=True,type_='shadow'))))
p3.render_notebook()
```

## 3.4 各省市4A-5A景区数量玫瑰图

```python
df_counts=df_counts.sort_values(ascending=False)
```

```python
p4=(Pie()
   .add('',[list(z) for z in zip(df_counts.index.values.tolist(),df_counts.values.tolist())],
       radius=['30%','100%'],
       rosetype='area')
   .set_global_opts(title_opts=opts.TitleOpts(title='地区景点数量'),
                   legend_opts=opts.LegendOpts(pos_left='2%',pos_top='12%',orient='vertical'),
                   toolbox_opts=opts.ToolboxOpts())
   .set_series_opts(label_opts=opts.LabelOpts(is_show=True,position='inside',font_size=12,
                                             formatter='{b}:{c} ({d}%)')))
p4.render_notebook()
```

## 3.5各省市4A-5A景区地图分布

```python
p5=(Map()
   .add('4A-5A景区在国内分布',[list(z) for z in zip(df_counts.index.values.tolist(),df_counts.values.tolist())],'china')
   .set_global_opts(title_opts=opts.TitleOpts(title='地图数据分布'),
                  visualmap_opts=opts.VisualMapOpts(max_=50,is_piecewise=True)))
p5.render_notebook()
```

```python
bins=[0,50,100,150,200,250,300,350,400,500]
label_level = ['0-50', '50-100', '100-150', '150-200', '200-250', '250-300', '300-350', '350-400', '400-500']
jz=pd.cut(df.价格,bins=bins,labels=label_level)
df_price=jz.value_counts()
df_price
```

## 3.6 景区门票价格区间占比玫瑰图

```python
p6=(Pie()
   .add('',[list(z) for z in zip(df_price.index.tolist(),df_price.values.tolist())],
       radius=['20%','50%'],
       rosetype='radius',
       label_opts=opts.LabelOpts(is_show=True))
   .set_global_opts(title_opts=opts.TitleOpts(title='门票价格占比',pos_left='center'),
                   legend_opts=opts.LegendOpts(type_='scroll',orient='vertical',pos_left='80%',pos_top='26%'),
                   )
   .set_series_opts(label_opts=opts.LabelOpts(position='outside',formatter='{b}:{c} ({d}%)')))
p6.render_notebook()
```

## 3.7门票价格区间数量散点图

```python
p7=(Scatter()
   .add_xaxis(df_price.index.tolist())
   .add_yaxis('门票区间价格',df_price.values.tolist())
   .set_global_opts(title_opts=opts.TitleOpts(title='门票价格区间数量'),
                   xaxis_opts=opts.AxisOpts(name='价格区间(元)'),
                   yaxis_opts=opts.AxisOpts(name='数量'),
                   visualmap_opts=opts.VisualMapOpts(type_='size',range_size=[5,50])))
p7.render_notebook()
```
