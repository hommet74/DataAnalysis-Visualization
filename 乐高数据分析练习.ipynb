```python
import pandas as pd
import numpy as np
import jieba 
from pyecharts.charts import Bar,Map,Pie
from pyecharts import options as opts
```

```python
df=pd.read_csv('./乐高淘宝数据.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df=df[df.purchase_num.str.contains('人付款')]
```

```python
df=df.reset_index(drop=True)
df.info()
```

```python
df.purchase_num=df.purchase_num.str.extract('(\d+)').astype('int')
```

```python
df.purchase_num
```

```python
df['sales_volume']=df.price*df.purchase_num
```

```python
df['province']=df.location.str.split(' ').str[0]
```

```python
df.head()
```

```python
shop_top10=df.groupby('shop_name')['purchase_num'].sum().sort_values(ascending=False).head(10)
```

```python
shop_top10
```

```python
bar=(Bar(init_opts=opts.InitOpts(theme='dark',width='800px'))
    .add_xaxis(shop_top10.index.tolist())
    .add_yaxis('',shop_top10.values.tolist())
    .set_global_opts(title_opts=opts.TitleOpts(title='乐高销量排名前十淘宝店铺'),
                    xaxis_opts=opts.AxisOpts(axislabel_opts=opts.LabelOpts(rotate=-8)),
                    visualmap_opts=opts.VisualMapOpts(max_=28669)))
bar.render_notebook()
```

```python
province_top10=df.province.value_counts()[:10]
```

```python
province_top10
```

```python
bar=(Bar(init_opts=opts.InitOpts(width='800px'))
    .add_xaxis(province_top10.index.tolist())
    .add_yaxis('',province_top10.values.tolist())
    .set_global_opts(title_opts=opts.TitleOpts(title='乐高产地排名前十'),
                    visualmap_opts=opts.VisualMapOpts(max_=1000,orient='horizontal')))
bar.render_notebook()
```

```python
province_num=df.groupby('province')['purchase_num'].sum().sort_values(ascending=False)
province_num
```

```python
map=(Map()
    .add('',[list(z) for z in zip(province_num.index.tolist(),province_num.values.tolist())],maptype='china')
    .set_global_opts(title_opts=opts.TitleOpts(title='乐高哥产地销量分布图'),
                    visualmap_opts=opts.VisualMapOpts(max_=172277)))
map.render_notebook()
```

```python
cut_bins=[0,50,100,200,300,500,1000,8888]
cut_labels=['0-50元','50-100元','100-200元','200-300元','300-500元','500-1000元','1000元以上']

price_cut=pd.cut(df.price,bins=cut_bins,labels=cut_labels)
price_num=price_cut.value_counts()
price_num
```

```python
bar=(Bar(init_opts=opts.InitOpts(theme='infographic',width='800px'))
    .add_xaxis(price_num.index.tolist())
    .add_yaxis('',price_num.values.tolist())
    .set_global_opts(title_opts=opts.TitleOpts(title='不同价格区间产品数量'),
                    visualmap_opts=opts.VisualMapOpts(max_=900)))
bar.render_notebook()
```

```python
df['price_cut']=price_cut
cut_purchase=df.groupby('price_cut')['sales_volume'].sum()
cut_purchase
```

```python
pie=(Pie()
    .add('',[list(z) for z in zip(cut_purchase.index.tolist(),cut_purchase.values.tolist())],radius=['35%','60%'])
    .set_global_opts(title_opts=opts.TitleOpts(title='不同价格区间销售额整体表现'),
                    legend_opts=opts.LegendOpts(orient='vertical',pos_top='15%',pos_left='5%'))
    .set_series_opts(label_opts=opts.LabelOpts(formatter='{b}:{d}%')))
pie.render_notebook()
```
