## 1.导入模块

```python
from pyecharts.charts import Map,Bar,Pie
from pyecharts import options as opts
import pandas as pd
```

## 2. Pandas 数据处理

### 2.1 读取数据

```python
df=pd.read_csv('./中国大学综合排名2021.csv',encoding='gb2312')
df.head()
```

```python
df.dtypes
```

```python
df.describe()
```

### 2.5 填充某一列缺失数据

```python
df['升/降'].fillna(0,inplace=True)
df.isnull().sum()
```

### 2.6一次性填充所有缺失数据

```python
df.fillna(0,inplace=True)
df.isnull().sum()
```

```python
df.describe()
```

### 2.7 统计所有排名未改变的学校

```python
df[df['升/降']==0]
```

```python
### 2.8 统计排名前50中排名下降的学校
```

```python
df.loc[(df['排名']<=50)&(df['升/降']<0),:]
```

### 2.9统计各个省市大学数量

```python
df_count=df.groupby('省市').agg(counts=('排名','count')).sort_values('counts',ascending=False)
df_count
```

### 2.10 各省市大学平均分排序

```python
df_means=df.groupby('省市').agg(平均分=('总分','mean'))
```

```python
df_means=round(df_means,2)
df_means=pd.concat([df_means,df_count],join='outer',axis=1).sort_values('平均分',ascending=False)
df_means
```

## 3. Pyecharts 数据可视化

### 3.1 各省市大学数量和平均分柱状图(横向)

```python
p1=(Bar()
   .add_xaxis(df_means.index.tolist())
   .add_yaxis('平均分',df_means['平均分'].values.tolist())
   .add_yaxis('大学数量',df_means['counts'].values.tolist())
   .set_global_opts(title_opts=opts.TitleOpts(title='中国大学排名',subtitle='各省市大学数量/平均分'),
                   xaxis_opts=opts.AxisOpts(name='省份'),
                   yaxis_opts=opts.AxisOpts(name='量'),
                   datazoom_opts=opts.DataZoomOpts(),
                   toolbox_opts=opts.ToolboxOpts()))

p1.render_notebook()
```

### 3.2 各省市大学数量玫瑰图

```python
p2=(Pie()
   .add('',[list(z) for z in zip(df_count.index.tolist(),df_count['counts'].values.tolist())],
       radius=['20%','60%'],
       center=['50%','60%'],
       rosetype='radius',
       label_opts=opts.LabelOpts(is_show=False))
   .set_global_opts(title_opts=opts.TitleOpts(title='各省市大学数量',subtitle='玫瑰图'),
                   legend_opts=opts.LegendOpts(pos_left='2%',pos_top='15%',orient='vertical'))
   .set_series_opts(label_opts=opts.LabelOpts(formatter='{b}:{c}',position='outer')))
p2.render_notebook()
```

### 3.3 各省市大学数量地图分布

```python
p3=(Map()
   .add('大学数量',[list(z) for z in zip(df_count.index.tolist(),df_count['counts'].values.tolist())],'china')
   .set_global_opts(visualmap_opts=opts.VisualMapOpts(max_=40,is_piecewise=True)))
p3.render_notebook()
```
