#### 2.2
通过食堂刷卡记录，分别绘制工作日和非工作日就餐时间曲线图，分析食堂早中晚餐的就餐峰值，并描述

```python
#创建一个消费星期列，周一为1，周七为7
data12['消费星期']=data12['消费时间'].dt.dayofweek+1
data12
```

- 以周一至周五划分为工作日，其余为非工作日


```python
work_day=data12.loc[:,'消费星期']<=5
unwork_day=data12.loc[:,'消费星期']>5

work_day=data12.loc[work_day,:]
unwork_day=data12.loc[unwork_day,:]
```

```python
work_day['小时']=work_day['消费时间'].apply(lambda x:str(x).split(" ")[-1].split(":")[0])
```

```python
work_day_time_count=work_day.groupby('小时')['流水号'].count()
```

```python
unwork_day['小时']=unwork_day['消费时间'].apply(lambda x:str(x).split(" ")[-1].split(":")[0])
```

```python
unwork_day_time_count=unwork_day.groupby('小时')['流水号'].count()
```

- 绘制工作日以及非工作日消费频次曲线图

```python
line1=(
    Line()
    .add_xaxis(work_day_time_count.index.tolist())
    .add_yaxis('工作日',work_day_time_count.values.tolist())
    .add_yaxis('非工作日',unwork_day_time_count.values.tolist())
    .set_global_opts(
        title_opts=opts.TitleOpts(title='工作日/非工作日消费曲线图',subtitle='2019-04-01至2019-04-30')
    )
)
line1.render_notebook()
```

#### 3.2根据学生的整体校园消费行为，选择合适的特征，构建聚类模型，分析每一类学生群体的消费特点

- 选择每个学生的性别，总消费金额，总消费次数为特征

```python
data_mo=data12.groupby('校园卡号').agg({'消费金额':sum,'消费次数':'count',"性别":'unique'})
```

```python
data_ce=data12
data_ce
```

```python
data_mo['性别']=data_mo['性别'].astype(str)
data_mo
```

```python
data_ce['性别']=data_ce['性别'].apply(lambda x: 1 if x='')
```

```python
data_mo['性别']=data_mo['性别'].apply(lambda x: 1 if x=='男' else 0)
```

```python
data_mo
```

```python
data_mo.columns=['总消费金额','总消费次数','性别']
```

```python
data_mo
```

- 构建聚类模型

```python
from sklearn.cluster import KMeans
data_zs=1.0*(data_mo-data_mo.mean())/data_mo.std()
model=KMeans(n_clusters=3,n_jobs=4,max_iter=500,random_state=10)
model.fit(data_zs)
```

```python
#r1统计各个类别的数目，r2找出聚类中心
r1=pd.Series(model.labels_).value_counts()
r2=pd.DataFrame(model.cluster_centers_)
r=pd.concat([r2,r1],axis=1)
r.columns=list(data_mo.columns)+['类别数目']
r
```

```python
r=pd.concat([data_mo,pd.Series(model.labels_,index=data_mo.index)],axis=1)
r.columns=list(data_mo.columns)+['聚类类别']
r.head()
```
