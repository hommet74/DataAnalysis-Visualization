### 用户销售贡献曲线

```python
s13=dt.groupby('user_id').agg(消费金额=('amount','sum'))
s13.sort_values('消费金额',ascending=False,inplace=True)
s13['累计消费总额']=s13.消费金额.cumsum()
s13['累计消费占比%']=s13.累计消费总额/s13.消费金额.sum()*100
s13['rank']=s13.消费金额.rank(ascending=False)
s13['rank']=s13['rank'].astype('int64')
s13['累计用户占比%']=s13['rank']/s13['rank'].max()*100
s13
```

```python
x=[round(x,2)for x in s13['累计用户占比%'][::1000]]
y=[round(x,2)for x in s13['累计消费占比%'][::1000]]

p13=(Line()
    .add_xaxis(x)
    .add_yaxis('前%客户贡献%销售额',y,
              symbol='emptyCircle',
              is_symbol_show=True,
              is_smooth=True,
              label_opts=opts.LabelOpts(is_show=False))
    .set_global_opts(tooltip_opts=opts.TooltipOpts(is_show=True),
                    xaxis_opts=opts.AxisOpts(type_='category',
                                            axistick_opts=opts.AxisTickOpts(is_show=True),
                                            splitline_opts=opts.SplitLineOpts(is_show=True)),
                    yaxis_opts=opts.AxisOpts(type_='value',
                                            axistick_opts=opts.AxisTickOpts(is_show=True),
                                            splitline_opts=opts.SplitLineOpts(is_show=True))))

p13.render_notebook()
```

### 标准归一化评分

```python
def score_c(x):
    score=(x.values-x.min())/(x.max()-x.min())
    return score
```

### Z-标准化

```python
def score_z(x):
    score=(x.values-x.mean())/x.std()
    return score
```

```python
s20=user_data.groupby('age_bin').agg(订单数量=('订单数量','sum'),订单总额=('订单总额','sum'),客单均价=('客单均价','mean'),客均单量=('订单数量','mean')).reset_index()
```

```python
s20
```

```python
score1=score_c(s20.订单数量)
score2=score_c(s20.订单总额)
score3=score_c(s20.客单均价)
score4=score_c(s20.客均单量)
s20['score']=score1+score2+score3+score4
s20['购买力rank']=s20['score'].rank(ascending=False)
s20.sort_values(by='购买力rank')
```

# RFM模型
R：最近一次消费（Recency）  
F：消费频率（Frequency）(按天)  
M：消费金额（Monetary）  

```python
rfm=dt.groupby(['user_id','date']).agg(消费金额=('amount','sum')).reset_index()
```

```python
rfm
```

```python
rfm=rfm.groupby('user_id').agg(最后消费日期=('date','max'),F=('date','count'),M=('消费金额','sum'))
```

```python
rfm
```

```python
rfm.info()
```

```python
rfm['最后消费日期']=pd.to_datetime(rfm.最后消费日期)
```

```python
rfm.info()
```

```python
rfm['R']=(rfm['最后消费日期'].apply(lambda x:rfm.最后消费日期.max()-x)).dt.days
```

```python
rfm.info()
```

```python
rfm
```

```python
rfm.describe(percentiles=(0.01,0.1,0.25,0.75,0.9,0.95,0.99))
```

```python
rfm2=rfm[['R','F','M']].reset_index(drop=True)
```

```python
rfm2
```

```python
def rfm_func(x):
    level=x.map(lambda x:'1'if x >=0 else '0')
    label=level.R+level.F+level.M
    d={
        '111':'重要价值客户',
        '011':'重要保持客户',
        '101':'重要挽留客户',
        '001':'重要发展客户',
        '110':'一般价值客户',
        '010':'一般保持客户',
        '100':'一般挽留客户',
        '000':'一般发展客户',
    }
    result=d[label]
    return result

rfm2['label']=rfm2.apply(lambda x:x-x.mean()).apply(rfm_func,axis=1)
rfm2
```

```python
rfm2['label'].value_counts()
```
